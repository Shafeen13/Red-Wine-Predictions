```python
# a mapping dictionary that maps the quality values from 0 to 5
quality_mapping = {
 3: 0,
 4: 1,
 5: 2,
 6: 3,
 7: 4,
 8: 5
}
wine.loc[:, "quality"] = wine.quality.map(quality_mapping)
```

```python
b = sb.countplot(x='quality', data=wine)
b.set_xlabel("quality", fontsize=10)
b.set_ylabel("count", fontsize=10)
```

```python
# Create Classification version of target variable
# 1 is good
# 0 is bad
wine['goodquality'] = [0 if x <= 2 else 1 for x in wine['quality']]# Separate feature variables and target variable
X = wine.drop(['quality','goodquality'], axis = 1)
Y = wine['goodquality']
```

```python
# analyzing the different values present in the dependent variable(quality column)
wine['goodquality'].value_counts()
```

```python
X.head()
```

```python
Y.head()
```

Feature Importance

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

from sklearn.ensemble import ExtraTreesClassifier
classifiern = ExtraTreesClassifier()
classifiern.fit(X,Y)
score = classifiern.feature_importances_
print(score)
```

Splitting Dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=1000, test_size=598, random_state=44)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

Using Decision Tree:

```python
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier(criterion='entropy',random_state=44)
model1.fit(x_train, y_train)
y_pred1 = model1.predict(x_test)

from sklearn.metrics import accuracy_score

print("Accuracy Score:", accuracy_score(y_test, y_pred1))
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
model2.fit(x_train,y_train)
y_pred2 = model2.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy Score:",accuracy_score(y_test,y_pred2))
```

```python
confusion_mat = confusion_matrix(y_test,y_pred2)
print(confusion_mat)
```

Using KNN:

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=3)
model3.fit(x_train,y_train)
y_pred3 = model3.predict(x_test)

from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(y_test,y_pred3))
```

Using SVC:

```python
from sklearn.svm import SVC
model4 = SVC()
model4.fit(x_train,y_train)
y_pred4 = model4.predict(x_test)

from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(y_test,y_pred4))
```

```python
results = pd.DataFrame({
    'Model': ['Logistic Regression','KNN', 'SVC','Decision Tree'],
    'Score': [0.752,0.650,0.630,0.722]})

result_wine = results.sort_values(by='Score', ascending=False)
result_wine = result_wine.set_index('Score')
result_wine
```

From the  above we can see that in Logistic Regression accuracy is greater than other  model so I will use Logistic Regression algorithm for training my model.
